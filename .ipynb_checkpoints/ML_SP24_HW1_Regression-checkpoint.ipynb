```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

***Type your code and answers below***

```python
# copy it to df variable for convenient
df = housing
df.shape
```

```python
# Type your code
#1.A check for null and outliers 
print(df.isnull().sum())
numerical_df = df.select_dtypes(include=['float64', 'int64'])
# a for loop to print boxplot and scatterplot for all numerical fields
for column in numerical_df.columns:
    sns.boxplot(x=numerical_df[column])
    plt.title(f'Boxplot for {column}')
    plt.show()
```

```python
##1.b Remove outliers and replacing null values
df = df[df['Income'] >= 35000]
df = df[df['HouseAge'] >= 2]
df = df[df['NumberOfRooms'] <= 20]

df['HouseAge'] = df['HouseAge'].fillna(df['HouseAge'].mean())
```

```python
##1.c Show that the outliers and null values are gone.
print(df.isnull().sum())
numerical_df = df.select_dtypes(include=['float64', 'int64'])
# a for loop to print boxplot and scatterplot for all numerical fields
for column in numerical_df.columns:
    sns.boxplot(x=numerical_df[column])
    plt.title(f'Boxplot for {column}')
    plt.show()
```

```python
##2.A correlation to identify the most important variable
hcorr = numerical_df.corr()
hcorr.style.background_gradient()

```

```python
##2.B Use Scatterplot ro identify the most important variable
sns.pairplot(numerical_df)
```

### The most important varaible is price.  

```python
###3.Build a one-factor Linear Regression model with that feature 

### i chose income and price


# 2. split samples
X= df['Income'].values
Y = df['Price'].values

X = X.reshape(-1, 1)
Y = Y.reshape(-1, 1)




X_train,X_test,y_train,y_test = train_test_split(X,
                                                 Y,
                                                 test_size=0.2,
                                               random_state=42)





lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
##4.Print out the model performance metrics (MSE, R-square). 
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score

print("bias is "+str(lr.intercept_))
print("coefficients  is "+str(lr.coef_) + "\n")

### using train data 
y_train_pred = lr.predict(X_train)

mae = mean_absolute_error(y_train, y_train_pred)
mse = mean_squared_error(y_train, y_train_pred)
rmse = np.sqrt(mse)
R_squared = r2_score(y_train, y_train_pred)
print('prediction for training set:')
print('MAE is: {}'.format(mae))
print('MSE is: {}'.format(mse))
print('RMSE is: {}'.format(rmse))
print('R_squared is: {}\n'.format(R_squared))

### using test data
y_test_pred = lr.predict(X_test)

mae = mean_absolute_error(y_test,y_test_pred)
mse = mean_squared_error(y_test, y_test_pred)
rmse = np.sqrt(mse)
R_squared = r2_score(y_test, y_test_pred)
print('prediction for testing set:')
print('MAE is: {}'.format(mae))
print('MSE is: {}'.format(mse))
print('RMSE is: {}'.format(rmse))
print('R_squared is: {}'.format(R_squared))
```

```python
###Plot your best-fitted line on top of the data. 

plt.scatter(X_test, y_test,  color='gray')
plt.scatter(X_test, y_test_pred, color='red', linewidth=2)
plt.show()
```

```python
##Perform a 10-fold cross validation and determine if your model pass the cross-validation.
import random
for i in range(10):
    print(random.randint(1, 1000))
    
for i in range(10):
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state = random.randint(1,1000))
    model = LinearRegression()
    model.fit(X_train, Y_train)
    Y_pred = model.predict(X_test)
    print('Trial {}, R-squared: {},  Slope: {}, Intercept: {},  '.format(i, metrics.r2_score(Y_test, Y_pred), \
                                                                      model.coef_[0], model.intercept_))
```

### My model pass the 10 fold validation, because we can r^2 is changing. The best trial is trail 8.

```python
# Load the dataset
df = pd.read_csv("data/affairs_modified.csv")
df.head()
```

```python
df.describe()
```

```python
df.isnull().any() 
```

```python
# we can simply drop them
df.dropna(inplace=True)
df.shape
```

Now, create a new column called 'had_affair'. We will set this column equal to 0 if the affairs column is 0 (meaning no time spent in affairs) otherwise the 'Had_Affair' value will be set as 1, indicating that the woman had an affair.

```python
# has_affair
def has_affair(x):
    if x != 0:
        return 1
    else:
        return 0

# Apply to DataFrame
df['had_affair'] = df['affairs'].apply(has_affair)
```

```python
df.tail()
```

```python
# Now we can drop the original affairs variable
df.drop(['affairs'], inplace=True, axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(numerical_df.drop('had_affair', axis=1), numerical_df['had_affair'], test_size=0.20, random_state=101)

from sklearn.preprocessing import StandardScaler
###normalize the data
normalizer = StandardScaler()
X_train = normalizer.fit_transform(X_train)
X_test = normalizer.transform(X_test)

model = LogisticRegression()
model.fit(X_train,y_train)

predictions = model.predict(X_test)
print(classification_report(y_test,predictions))
print(accuracy_score(y_test, predictions))
```

```python
### 5. Use one-hot encoding to include the categorical variable to build a second model.
```

```python
from sklearn.preprocessing import OneHotEncoder
cat_encoder = OneHotEncoder()
cat_1hot = cat_encoder.fit_transform(df[['religious', 'educ', 'occupation','occupation_husb' ]])
```

```python
cat_1hot.toarray()
```

```python
cat_encoder.categories_
```

```python
cat_1hot_df = pd.DataFrame(cat_1hot.toarray(), columns = ['fairly', 'midly', 'not', 'strongly',
'advanced degree', 'college', 'grade school', 'high school',
        'some college', 'some graduate school',
'occ1', 'occ2', 'occ3', 'occ4', 'occ5', 'occ6',
'husb_occ1', 'husb_occ2', 'husb_occ3', 'husb_occ4', 'husb_occ5',
        'husb_occ6'])
cat_1hot_df.head()
```

```python
full_df = pd.concat([numerical_df, cat_1hot_df], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(full_df.drop('had_affair', axis=1), full_df['had_affair'], test_size=0.30, random_state=101)

model = LogisticRegression()
model.fit(X_train,y_train)
predictions = model.predict(X_test)
print(classification_report(y_test,predictions))
print(accuracy_score(y_test, predictions))
```
